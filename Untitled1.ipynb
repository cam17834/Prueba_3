```python
import random

Datos_Personas=[]

def validar_rut(rut):
  rut = rut.replace(".", "").replace("-", "")
  if len(rut) != 9 or not rut[:-1].isdigit():
    return False
  multiplicadores = [2, 3, 4, 5, 6, 7, 2, 3]
  total = sum(int(digito) * multiplicador for digito, multiplicador in zip(rut[:-1][::-1], multiplicadores))
  resto = total % 11
  digito_verificador = str(11 - resto)
  if digito_verificador == "11":
    digito_verificador = "0"
  elif digito_verificador == "10":
    digito_verificador = "K"
  return rut[-1].upper() == digito_verificador


def grabar():
  rut=input("Ingrese  Rut:")
  print("*_________________________________________________________________________*")
  while not validar_rut(rut):
    print("ERROR!  Rut ingresado no esta registrado o es incorrecto.")
    print("*_______________________________________________________________________*")
    rut=input("Ingrese Rut: ")
    print("*_______________________________________________________________________*")

  nombre=input("Ingrese Nombre Completo: ")
  print("*_________________________________________________________________________*")
  while len(nombre)<8:
      print("NOMBRE INVALIDO, (Debe contener al menos 8 caracteres).")
      nombre=input('Ingrese Nombre Completo: ')
      print("*_____________________________________________________________________*")


  edad=int(input("Ingrese Edad: "))
  print("*_________________________________________________________________________*")
  while edad<=0 or edad>=120:
      print("Ingrese edad real. (entre 0 y 120)")
      print("*_____________________________________________________________________*")
      edad=int(input("\nIngrese Edad: "))
      print("*_____________________________________________________________________*")

  fech_nac = None

  while fech_nac is None or not fech_nac:
      fech_nac = input("Ingrese Fecha de Nacimiento (DD/MM/AAAA): ")
      print("*_____________________________________________________________________*")



  while True:
      partes = fech_nac.split("/")
      if len(partes) != 3 or not all(p.isdigit() for p in partes):
          print("*_________________________________________________________________*")
          print("Dato Incorrecto.")
          print("Vuelva a intentar porfavor!")
          print("*_________________________________________________________________*")
          fech_nac = input("Ingrese fecha de nacimiento (DD/MM/AAAA): ")

      else:
          dia, mes, año = map(int, partes)
          if 1 <= dia <= 31 and 1 <= mes <= 12 and 1900 <= año <= 2023:
              break
          else:
              print("Fecha ingresada No Válida")
              fech_nac = input("Vuelva a intentar porfavor!")

  estado_civil=input('Ingrese Estado Civil (Soltero/a - Casado/a - Divorciado/a - Viudo/a): ').lower()
  print("*_________________________________________________________________________*")
  while not estado_civil=="soltero" and estado_civil=="soltera" and estado_civil=="soltero/a" and estado_civil=="casado" and estado_civil=="casada" and estado_civil=="casado/a" and estado_civil=="divorciado" and estado_civil=="divorciada" and estado_civil=="divorciado/a" and estado_civil=="viudo" and estado_civil=="viuda" and estado_civil=="viudo/a":
      print("Dato ingresado NO válido. ")
      print("Vuelva a intentar porfavor! \n(Soltero/a - Viudo/a - Separado/a - Divorciado/a)")
      print("*_____________________________________________________________________*")
      estado_civil=input("Ingrese Estado Civil (Soltero/a - Casado/a - Divorciado/a - Viudo/a): ").lower()
      print("*_____________________________________________________________________*")

  pais=input("Ingrese País de origen: ").lower()
  print("*_________________________________________________________________________*")
  while not pais.isalpha():
      print("Debe contener sólo letras")
      print("*_____________________________________________________________________*")
      pais=input("\nIngrese País de origen:").lower()
      print("*_____________________________________________________________________*")
  ciudad=input("Ingrese Ciudad de origen:").lower()
  print("*_________________________________________________________________________*")
  while not ciudad.isalpha():
      print("Debe conetener sólo letras")
      print("*_____________________________________________________________________*")
      ciudad=input("\nIngrese su ciudad de origen:").lower()
      print("*_____________________________________________________________________*")

  persona={
      "rut":rut,
      "nombre":nombre,
      "edad":edad,
      "fecha_de_nacimiento": {"dia":dia,"mes":mes,"año":año },
      "pais_nacimiento":pais,
      "ciudad_nacimiento":ciudad,
      "estado_civil":estado_civil }
  datos_personas.append(persona)
  print("\nDatos Registrados")

def buscar():
    rut=input("Ingrese Rut: ")
    print("*_______________________________________________________________________*")
    while not validar_rut(rut):
      print("ERROR: El rut ingresado es erroneo o no esta registrado.")
      print("*_____________________________________________________________________*")
      rut=input("Ingrese Rut: ")
      print("*_____________________________________________________________________*")
    for persona in datos_personas:
      if persona["rut"]==rut:
          print("")
          print("Información de la persona:")
          print("Nombre: ", persona["nombre"])
          print("RUT: ", persona["rut"])
          print("Fecha de nacimiento: {}/{}/{}".format(
              persona["fecha_de_nacimiento"]["dia"],
              persona["fecha_de_nacimiento"]["mes"],
              persona["fecha_de_nacimiento"]["año"]))
          print("Edad: ", persona["edad"])
          print("Estado civil: ", persona["estado_civil"])
          print("Pais de nacimiento:", persona["pais_nacimiento"])
          print("Ciudad nacimiento:", persona["ciudad_nacimiento"])
          if persona["pais_nacimiento"]=="chile":
              print("Nacionalidad Chilena.")
          else:
              print("No posee Nacionalidad Chilena")
      else:
        print("Dato Registrado No Válido")


def imprimir_cert():
  print("*_________________________________________________________________________*")
  rut=input("Ingrese Rut: ")
  print("*_________________________________________________________________________*")
  while not validar_rut(rut):
    print("ERROR: El rut ingresado es erroneo o no esta registrado.")
    print("*_______________________________________________________________________*")
    rut=input("Ingrese Rut: ")
    print("*_______________________________________________________________________*")

  persona_encontradaf = False

  for persona in datos_personas:
    if persona["rut"] == rut:
        persona_encontrada = persona
        break

  if persona_encontrada:
      print("Certificados  Personas con RUT:", persona_encontrada["rut"])
      print("Nombre:", persona_encontrada["nombre"])
      persona_encontradaf = True

  if persona_encontradaf:
    while True:
      print("\nTipos de Certificados:")
      print("_1._ Certificado de Nacimiento")
      print("_2._ Certificado de Estado Civil")
      print("_3._ Certificado de Pertenencia")
      print("_4._ Volver al Menú Principal")

      print("*_____________________________________________________________________*")
      opcion2 = int(input("Elija una opción"))
      print("*_____________________________________________________________________*")
      if opcion2==1:
        print('')
        print("*________________*-*-*-*-*-*-*-*-*______________________*")
        print("      ===== CERTIFICADO DE NACIMIENTO =====        ")
        print("*________________*-*-*-*-*-*-*-*-*______________________*")
        print("CIRCUNSCRIPCIÓN:\t",persona["ciudad_nacimiento"].upper())
        print("AÑO DE INSCRIPCIÓN:\t",persona['fecha_de_nacimiento']['año'])
        print("NOMBRE:\t\t\t",persona['nombre'].upper())
        print("RUN:\t\t\t",persona["rut"])
        print("*________________*-*-*-*-*-*-*-*-*______________________*")


      elif opcion2==2:
        print("")
        print("*________________*-*-*-*-*-*-*-*-*______________________*")
        print("        ===== CERTIFICADO DE ESTADO CIVIL =====         ")
        print("*________________*-*-*-*-*-*-*-*-*______________________*")
        print("NOMBRE:\t\t\t",persona["nombre"].upper())
        print("RUN:\t\t\t",persona["rut"])
        print("ESTADO CIVIL:\t\t",persona["estado_civil"].upper())
        print("*________________*-*-*-*-*-*-*-*-*______________________*")


      elif opcion2==3:
        print("")
        print("*________________*-*-*-*-*-*-*-*-*______________________*")
        print("       ===== CERTIFICADO DE PERTENENCIA =====        ")
        print("*________________*-*-*-*-*-*-*-*-*______________________*")
        print("NOMBRE:\t\t\t",persona["nombre"].upper())
        print("RUN:\t\t\t",rut)
        print("CIUDAD DE NACIMIENTO:\t",persona["ciudad_nacimiento"].upper())
        print("PAIS DE NACIMIENTO:\t", persona["pais_nacimiento"].upper())
        print("")
        print("SE CERTIFICA QUE EL/LA SEÑOR/A")
        print("")
        print("NOMBRE:\t\t\t",persona["nombre"].upper(),)
        print("RUN:\t\t\t",persona["rut"])
        print("PERTENECE A:\t\t",persona["pais_nacimiento"].upper())
        print("*________________*-*-*-*-*-*-*-*-*______________________*")


      elif opcion2==4:

        print('Volviendo al Menú principal.')
        print("")
        break

      else:
        print("OPCIÓN INCORRECTA")


def eliminar():
    rut=input("Ingrese Rut:")
    print("*_________________________________________________________________________*")
    while not validar_rut(rut):
      print("ERROR: El rut ingresado es erroneo o no esta registrado.")
      print("*_________________________________________________________________________*")
      rut=input('Ingresa el rut: ')
      print("*_________________________________________________________________________*")

    for i, persona in enumerate(datos_personas):
        if persona["rut"] == rut:
            print("Registro Eliminado Exitósamente")
            print("*_________________________________________________________________________*")
            persona_eliminada = datos_personas.pop(i)
            break


def mostrar():
  cant_personas=0
  cant_personas=len(datos_personas)
  print(f"Personas registradas: {cant_personas}")
  for persona in datos_personas:
      print("")
      print("*________________*-*-*-*-*-*-*-*-*______________________*")
      print("Información de la persona:")
      print("Nombre: ", persona["nombre"])
      print("RUT: ", persona["rut"])
      print("Fecha de nacimiento: {}/{}/{}".format(
          persona["fecha_de_nacimiento"]["dia"],
          persona["fecha_de_nacimiento"]["mes"],
          persona["fecha_de_nacimiento"]["año"]))
      print("Edad: ", persona["edad"])
      print("Pais de nacimiento:", persona["pais_nacimiento"])
      print("Ciudad nacimiento:", persona["ciudad_nacimiento"])
      print("*________________*-*-*-*-*-*-*-*-*______________________*")


def imprimir_menu():
  print("*_________________________________________________________________________*")
  print("")
  print(" MENÚ :")
  print("1._ Guardar Datos")
  print("2._ Buscar")
  print("3._ Imprimir certificados")
  print("4._ Eliminar")
  print("5._ Mostrar Datos Registrados")
  print("6._ Salir")
  print('*_________________________________________________________________________*')

def main():
  print("Hola")
  print("Bienvenido a nuestro Totem Registro Civil")

if __name__ == "__main__":
  main()
  while True:
    imprimir_menu()
    print("*_________________________________________________________________________*")
    opcion = input("Elija una opción: ")
    print("*_________________________________________________________________________*")
    if opcion == "1":
      grabar()
    elif opcion == "2":
      buscar()
    elif opcion == "3":
      imprimir_cert()
    elif opcion == "4":
      eliminar()
    elif opcion == "5":
      mostrar()
    elif opcion == "6":
      print("Que tenga buen día...")
      print("Vuelva Pronto!.")
      print("*_________________________________________________________________________*")
      print('')

      break
    else:
      print("Opción Inválida.")
      print("Vuelva a intentar porfavor!")
```
